```python
import matplotlib.pyplot as plt 
import sys
sys.path.append("../src/")
from functions import *
```

```python
data = getStockDataVec('^GSPC')
# buy = float('nan') * 252
# sell = float('nan') * 252

buy = [0] * 252
sell = [0] * 252

for i in range(252):
    if i % 50 == 0: 
        buy[i] = 1
    
    elif i % 47 == 0:
        sell[i] = 1
        
buy = [buy[i] * data[i] for i in range(252)]
sell = [sell[i] * data[i] for i in range(252)]

for i in range(252):
    if buy[i] == 0: 
        buy[i] = float('nan')
        
    if sell[i] == 0: 
        sell[i] = float('nan')

```

```python
days = list(range(252))
plt.plot(days, data, 'b',
         days, buy, 'go',
         days, sell, 'rx')
plt.xlabel('Days')
plt.ylabel('$$$')
plt.title('S&P 500 2018')
plt.legend(['Price','buy','sell'])
```
